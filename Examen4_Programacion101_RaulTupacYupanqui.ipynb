```python
import pandas as pd
from google.colab import files

# Pregunta 1
# --- 1. Carga diabetes.csv en un DataFrame df_diabetes ---
print("Por favor, sube el archivo 'diabetes.csv':")
uploaded = files.upload()

# Asegúrate de que el archivo 'diabetes.csv' ha sido subido.
if 'diabetes.csv' not in uploaded:
    print("Error: 'diabetes.csv' no fue encontrado. Asegúrate de haberlo subido.")
else:
    try:
        df_diabetes = pd.read_csv('diabetes.csv')
        print("\n'diabetes.csv' cargado con éxito en df_diabetes.")

        # --- 2. Muestra las 5 primeras filas. ---
        print("\n--- Primeras 5 filas del DataFrame df_diabetes ---")
        print(df_diabetes.head())

        # --- 3. Imprime el número total de filas, columnas y el tipo de cada columna. ---
        print("\n--- Información general del DataFrame df_diabetes ---")
        print(f"Número de filas: {df_diabetes.shape[0]}")
        print(f"Número de columnas: {df_diabetes.shape[1]}")
        print("\nTipos de datos de cada columna:")
        df_diabetes.info() # df.info() ya muestra las filas, columnas y tipos de datos

        # --- 4. Convierte la columna Outcome a tipo categórico y verifica el cambio ---
        print("\n--- Convirtiendo la columna 'Outcome' a tipo categórico ---")
        df_diabetes['Outcome'] = df_diabetes['Outcome'].astype('category')

        print("\n--- Información del DataFrame df_diabetes después de la conversión de 'Outcome' ---")
        df_diabetes.info()

    except Exception as e:
        print(f"\nOcurrió un error al procesar 'diabetes.csv': {e}")


```

```python
# 1. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120.
print("\n--- 1. Creando df_filtrado (Glucose > 120) ---")
df_filtrado = df_diabetes[df_diabetes['Glucose'] > 120].copy()
print(f"Número de filas en df_filtrado: {df_filtrado.shape[0]}")
print("Primeras 5 filas de df_filtrado (solo para verificación inicial):")
print(df_filtrado.head())


# 2. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI así:
#    "Bajo" si BMI < 18.5
#    "Normal" si 18.5 <= BMI < 25
#    "Sobrepeso" si 25 <= BMI < 30
#    "Obeso" si BMI >= 30
print("\n--- 2. Agregando la columna BMI_Category ---")

# Definimos las condiciones y las etiquetas
conditions = [
    df_filtrado['BMI'] < 18.5,
    (df_filtrado['BMI'] >= 18.5) & (df_filtrado['BMI'] < 25),
    (df_filtrado['BMI'] >= 25) & (df_filtrado['BMI'] < 30),
    df_filtrado['BMI'] >= 30
]
categories = ['Bajo', 'Normal', 'Sobrepeso', 'Obeso']

# Usamos np.select para aplicar las condiciones
import numpy as np # Asegúrate de que numpy esté importado para np.select
df_filtrado['BMI_Category'] = np.select(conditions, categories, default='Desconocido') # default es opcional

print("Columna 'BMI_Category' creada.")

# También podemos verificar la distribución de las nuevas categorías
print("\nConteo de valores en 'BMI_Category':")
print(df_filtrado['BMI_Category'].value_counts())


# 3. Muestra las primeras 5 filas de df_filtrado con la nueva columna incluida.
print("\n--- 3. Primeras 5 filas de df_filtrado con la nueva columna BMI_Category ---")
print(df_filtrado.head())
```

```python
#Pregunta 3
import pandas as pd
from google.colab import files # Necesario para subir el segundo archivo

# 1. Carga age_groups.csv en un DataFrame df_age_groups.
print("\n--- 1. Subiendo y cargando 'age_groups.csv' ---")
print("Por favor, sube el archivo 'age_groups.csv':")
uploaded_age_groups = files.upload()

if 'age_groups.csv' not in uploaded_age_groups:
    print("Error: 'age_groups.csv' no fue encontrado. Asegúrate de haberlo subido.")
else:
    try:
        df_age_groups = pd.read_csv('age_groups.csv')
        print("\n'age_groups.csv' cargado con éxito en df_age_groups.")
        print("Primeras 5 filas de df_age_groups:")
        print(df_age_groups.head())
        print("\nInformación de df_age_groups:")
        df_age_groups.info()

        # 2. Haz un merge de df_diabetes con df_age_groups por la columna Age.
        #    El resultado se llamará df_combinado. Debe ser un left merge para mantener todas las filas de df_diabetes.
        print("\n--- 2. Realizando el left merge de df_diabetes con df_age_groups ---")
        df_combinado = pd.merge(df_diabetes, df_age_groups, on='Age', how='left')

        print("\nMerge completado. Número de filas en df_combinado:", df_combinado.shape[0])
        print("Número de columnas en df_combinado:", df_combinado.shape[1])


        # 3. Muestra las 5 primeras filas de df_combinado para verificar que la columna AgeGroup se haya agregado correctamente.
        print("\n--- 3. Primeras 5 filas de df_combinado con la columna AgeGroup ---")
        print(df_combinado.head())

        # Opcional: Verificar la información del DataFrame combinado para ver la nueva columna y tipos de datos
        print("\n--- Información de df_combinado después del merge ---")
        df_combinado.info()

    except Exception as e:
        print(f"\nOcurrió un error al procesar 'age_groups.csv' o al realizar el merge: {e}")
```
